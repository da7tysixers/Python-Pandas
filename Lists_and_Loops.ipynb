```python
acronyms = []

acronyms.append("NASA")
acronyms.append("FBI")
acronyms.append("CIA")
acronyms.append("UN")

print(acronyms) 

# Delete the first element of the list
del acronyms[3]
print(acronyms) 
```

```python
# Checking a list for an element
if "FBI" in acronyms:
    print("FBI is in the list")
if "NASA" in acronyms:
    print("NASA is in the list")
```

```python
for n in acronyms:
    print(n)
```

```python
# Expenses calculation
expenses = [10.50, 8, 5, 15, 20, 5, 3]
total = 0
for expense in expenses:
    total += expense
print(f"Total expenses for the week: {total}")

# printing the total using a separator and a comma
print("You spent $", total, sep='')

# Using the sum function to calculate the total expenses
print("You spent $", sum(expenses), sep='')
```

```python
expenses = []

my_expense = input("Enter an expense for the week: \n")
for i in range(len(my_expense.split(','))):
    num = my_expense.split(',')
    expenses.append(float(num[i]))
print("You spent $", sum(expenses), sep='')
```

```python
# Understading range in loops
# Range have 3 parameters: start, stop, step
# start is the number to start counting from (default is 0)
# stop is the number to stop counting at (not included in the range)
# step is the number to increment by (default is 1)
for i in range(5):
    print(i)
```
