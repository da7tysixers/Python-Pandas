```python
temp = 98
if temp == 95:
    status = "humid"
    print("It's hot outside!")
    print("Drink water and stay cool!")
    if status == "humid":
        print("Make sure to stay hydrated!")
elif temp < 95 or temp > 85:
    print("It's not too hot outside!")
    print("Enjoy the weather!")
else:
    print("It's really hot outside!")
    print("Stay indoors and keep cool!")
```

```python
import random
computer_choice = ['scissors', 'rock', 'paper']
user_choice = str(input('Enter rock, paper, or scissors: \n'))

if random.choice(computer_choice) == user_choice:
    if computer_choice == user_choice:
        print("It's a tie!")
        print("Computer chose:", random.choice(computer_choice))
        print("User chose:", user_choice)
        print(computer_choice)
    else:
        print("You win!")
        print("Computer chose:", random.choice(computer_choice))
        print("User chose:", user_choice)
else:
    print("Invalid choice. Please enter rock, paper, or scissors.")
    print("Computer chose:", random.choice(computer_choice))
    print("User chose:", user_choice)
```
